# Star Wars Survey Analysis

For this project, we'll be cleaning and analyzing a dataset created by [FiveThirtyeight](https://fivethirtyeight.com/) for their article, [America’s Favorite ‘Star Wars’ Movies (And Least Favorite Characters)](https://fivethirtyeight.com/features/americas-favorite-star-wars-movies-and-least-favorite-characters/). The data consists of 835 responses to a survey created by the FiveThirtyEight staff on SurveyMonkey, and presents some particularly difficult cleaning issues due to several questions that required the respondent to check one or more boxes. In this dataset, each of the potential responses becomes it's own column, requiring a fair amount of cleaning to sort out.

The original dataset can be found on FiveThirtyEight's Github, [here](https://github.com/fivethirtyeight/data/tree/master/star-wars-survey).




```python
#import libraries and dataset
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

star_wars = pd.read_csv('star_wars.csv', encoding='ISO-8859-1')
```

```python
pd.options.display.max_columns = None
pd.options.display.max_rows = None

#explore columns
star_wars.head()
```

## Columns

The first thing that immediately presents itself is that although there conventional columns such as *'Have you seen any of the 6 films in the Star Wars franchise?'* and *'Do you consider yourself to be a fan of the Star Wars film franchise?'*, there are three series which begin with a question, then have several *Unnamed* columns. 

- The first series begins with a column called *'Which of the following Star Wars films have you seen? Please select all that apply.'* which both represents the question and the first film. It is followed by five of the unnamed columns representing the other five films. 

- The second series begins with column *'Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.'* and follows the same format as the first. 

- The final series asks the question, *'Please state whether you view the following characters favorably, unfavorably, or are unfamiliar with him/her.* and is again followed by multiple unnamed columns representing different characters in the Star Wars universe.  

Each of these will have to be dealt with individually to pull meaningful data from it. We'll first remove those rows without valid respondent IDs, then we'll delve into the column issues.  


```python
#remove rows without a valid respondent ID
star_wars = star_wars[star_wars['RespondentID'].notnull()]
```

```python
#ensure the rows have been removed
star_wars.head()
```

```python
#create a map to rename the character columns
char_names = {
    'Please state whether you view the following characters favorably, unfavorably, or are unfamiliar with him/her.':'Han Solo',
    'Unnamed: 16': 'Luke Skywalker',
    'Unnamed: 17': 'Princess Leia Organa',
    'Unnamed: 18': 'Anakin Skywalker',
    'Unnamed: 19': 'Obi Wan Kenobi',
    'Unnamed: 20': 'Emperor Palpatine',
    'Unnamed: 21': 'Darth Vader',
    'Unnamed: 22': 'Lando Calrissian',
    'Unnamed: 23': 'Boba Fett',
    'Unnamed: 24': 'C-3P0',
    'Unnamed: 25': 'R2-D2',
    'Unnamed: 26': 'JarJar Binks',
    'Unnamed: 27': 'Padme Amidala',
    'Unnamed: 28': 'Yoda'
    
}

#rename the character columns 
star_wars = star_wars.rename(columns=char_names)
```

```python
#convert columns to boolean values for easier analysis
yes_no = {
    'Yes': True,
    'No': False,
    np.NaN: False
}

star_wars['Have you seen any of the 6 films in the Star Wars franchise?'] = star_wars['Have you seen any of the 6 films in the Star Wars franchise?'].map(yes_no)
star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'] = star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'].map(yes_no)
```

```python
#change the columns names for the 'Have you seen any...' question
#to make them more inituitive
star_wars = star_wars.rename(columns={
    "Which of the following Star Wars films have you seen? Please select all that apply.": "seen_1", 
    'Unnamed: 4':'seen_2',
    'Unnamed: 5':'seen_3',
    'Unnamed: 6':'seen_4',
    'Unnamed: 7':'seen_5',
    'Unnamed: 8':'seen_6'
})

#convert entries to boolean values
yes_no = {
    'Star Wars: Episode I  The Phantom Menace': True,
    'Star Wars: Episode II  Attack of the Clones': True,
    'Star Wars: Episode III  Revenge of the Sith': True,
    'Star Wars: Episode IV  A New Hope': True,
    'Star Wars: Episode V The Empire Strikes Back': True,
    'Star Wars: Episode VI Return of the Jedi': True,
    np.nan:False,
}

#map yes_no to the seen columns
for i in range(1,7):
    i=str(i)
    star_wars['seen_' + i] = star_wars['seen_' + i].map(yes_no)
```

```python
#convert columns to numeric type
star_wars.iloc[:,9:15] = star_wars.iloc[:,9:15].astype(float)
```

```python
#rename ranking columns
star_wars = star_wars.rename(columns={
    'Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.':'rank_ep_1',
    'Unnamed: 10': 'rank_ep_2',
    'Unnamed: 11': 'rank_ep_3',
    'Unnamed: 12': 'rank_ep_4',
    'Unnamed: 13': 'rank_ep_5',
    'Unnamed: 14': 'rank_ep_6'
    
})
```

```python
#Examine the cleaned dataset
star_wars.head()
```

# Analysis

Now that we've done some cleaning to make the data easier to work with, we'll start to analyze some trends around the Star Wars universe.  

```python
names_map = ['The Phantom Menace', 'Attack of the Clones', 'Revenge of the Sith',
            'A New Hope', 'The Empire Strikes Back', 'Return of the Jedi']

#find the mean of each rank column and display as a bar chart
rankings = star_wars.iloc[:,9:15].mean()
rank_chart = rankings.plot.bar()
rank_chart.set_title('Average Rank of Star Wars Movies')
rank_chart.set_ylabel('Average Score (lower is better)')
rank_chart.set_xlabel('Movies (Numbers correspond to episode numbers)')
rank_chart.set_xticklabels(labels=names_map)
plt.show()

#Find the sum of respondents who have seen each film and
#display as a bar chart
seen_cols = star_wars.iloc[:,3:9].sum()
films_seen = seen_cols.plot.bar()
films_seen.set_title('Films Seen')
films_seen.set_ylabel('Number of respondents Who have Seen the Film')
films_seen.set_xlabel('Movies (Numbers correspond to episode numbers)')
films_seen.set_xticklabels(labels=names_map)
plt.show()
```

# Movie Rankings
The columns renamed in cell 9 correspond to the question, *'Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.'*  where each column is the rank for one of the movies. In the bar chart above, I've compared the mean of each movie's rankings to see which is, on average, the most popular. The column titles correspond to the the episode numbers, so rank_ep_1 represents *Star Wars: Episode I - The Phantom Menace,* and so on.  

A couple of things stand out immediately. First, with an average score of about 2.5, *The Empire Strikes Back* is the highest ranked Star Wars film. It is also clear that the first trilogy (represented by rank_ep_4, 5, and 6) is consistently rated higher than the prequels. Second, the scores for each of the prequel movies get progressively worse with the majority of respondents ranking *Revenge of the Sith* as their least favorite film. 

Comparing the films' rankings with the number of respondents who have seen each film, we can see that more people have seen the original trilogy, which might account for part of their consistently higher ratings. However, the difference in people who have seen first trilogy and the people who have seen the prequels does not seem to be great enough to account for the different in ratings.

# Gender Differences


```python
#divide respondents by gender
males = star_wars[star_wars['Gender'] == 'Male']
females = star_wars[star_wars['Gender'] == 'Female']

#find the mean of the ranking by gender
male_rankings = males.iloc[:,9:15].mean()
female_rankings = females.iloc[:,9:15].mean()

#respondents who have seen the films by gender
male_seen_cols = males.iloc[:,3:9].sum()
female_seen_cols = females.iloc[:,3:9].sum()

```

```python
names_map = ['The Phantom Menace', 'Attack of the Clones', 'Revenge of the Sith',
            'A New Hope', 'The Empire Strikes Back', 'Return of the Jedi']

#create a bar plot of the rankings by gender
fig, ax = plt.subplots()
bar_positions = np.arange(len(names_map))
width = 0.35
men_rankings = ax.bar(bar_positions - width/2, male_rankings, width, label='Men')
women_rankings = ax.bar(bar_positions + width/2, female_rankings, width, label='Women')
ax.set_title('Film Rankings by Gender')
ax.set_ylabel('Average Rank')
ax.legend()
ax.set_xticks(bar_positions)
ax.set_xticklabels(names_map, rotation = 90)
plt.show()

#create a bar plot of the movies seen by gender
fig, ax = plt.subplots()
bar_positions = np.arange(len(names_map))
width = 0.35
men_movies_seen = ax.bar(bar_positions - width/2, male_seen_cols, width, label='Men')
women_movies_seen = ax.bar(bar_positions + width/2, female_seen_cols, width, label='Women')
ax.set_title('Films Seen by Gender')
ax.set_ylabel('Respondents Who\'ve Seen the Film')
ax.legend()
ax.set_xticks(bar_positions)
ax.set_xticklabels(names_map, rotation = 90)
plt.show()
```

Plotting the highest-ranked movies by gender shows us that although there are differences by gender, as a whole, the trends remain consistent: *The Empire Strikes Back* and *Revenge of the Sith* are the favorite and least-favorite films, respectively, for both genders (although the differences between the prequel films is less pronounced for men), and the original trilogy is also the clear favorite for both genders.  

Looking at which movies have been seen by gender, we can see that the number of men who have seen the movies is consistenly higher than the number of women, and this is particularly true for the prequels. 

# Fan Favorites

Next, we'll look at whether or not the rankings and films seen change based on whether someone considers themselves to be a fan of the series or not.  First, we'll rename the relevant column to make it somewhat easier to work with.

```python
#rename the fan column
star_wars = star_wars.rename(columns={'Do you consider yourself to be a fan of the Star Wars film franchise?': 'film_fan'})
```

```python
#isolate the fans from the rest of the respondents
fan = star_wars[star_wars['film_fan'] == True]
no_fan = star_wars[star_wars['film_fan'] == False]
```

```python
fan.head()
```

```python
no_fan.head()
```

```python
#isolate fan and non-fan rankings
fan_rankings = fan.iloc[:,9:15].mean()
no_fan_rankings = no_fan.iloc[:,9:15].mean()

#isolate fan and non-fan movies seen
fan_movies_seen = fan.iloc[:,3:9].sum()
no_fan_movies_seen = no_fan.iloc[:,3:9].sum()

#plot the difference in rankings based on whether the respondent considers themselves a fan 
fig, ax = plt.subplots()
bar_positions = np.arange(len(names_map))
width = 0.35
fan_rank = ax.bar(bar_positions - width/2, fan_rankings, width, label='Fan')
no_fan_rank = ax.bar(bar_positions + width/2, no_fan_rankings, width, label='Not a Fan')
ax.set_title('Film Rankings')
ax.set_ylabel('Average Rank')
ax.legend()
ax.set_xticks(bar_positions)
ax.set_xticklabels(names_map, rotation = 90)
plt.show()

#plot the differences in movies seen based on whether the respondents considers themselves a fan
fig, ax = plt.subplots()
bar_positions = np.arange(len(names_map))
width = 0.35
fans_movies = ax.bar(bar_positions - width/2, fan_movies_seen, width, label='Fan')
no_fan_movies = ax.bar(bar_positions + width/2, no_fan_movies_seen, width, label='Not a Fan')
ax.set_title('Films Seen')
ax.set_ylabel('Respondents Who\'ve Seen the Film')
ax.legend()
ax.set_xticks(bar_positions)
ax.set_xticklabels(names_map, rotation = 90)
plt.show()
```

Unsurprisingly, many more people who identify as fans of the series have seen the films than those who do not identify as fans.  More interesting is that fans rank the original trilogy higher than non-fans (again, remember lower is better), but the opposite holds true for the prequels. Let's see where else fans and non-fans differ.  


## Favorite Characters

The last thing we'll look at is how the different characters are viewed by both fans and non-fans. Because there are far fewer non-fan respondents, comparing absolute numbers wouldn't make much sense. Instead, we'll calcualte the percentage of respondents in each group who rated each character 'very favorably'. You could, perhaps, glean some extra insight by comparing each of the ratings for fans and non-fans, but for the sake of simplicity, we'll stick with which characters are viewed favorably.

```python
#initialize list of characters
char_list = ['Han Solo', 'Luke Skywalker', 'Princess Leia Organa', 'Anakin Skywalker', 'Obi Wan Kenobi',
             'Emperor Palpatine', 'Darth Vader', 'Lando Calrissian', 'Boba Fett', 'C-3P0', 'R2-D2', 'JarJar Binks',
             'Padme Amidala', 'Yoda']

#initialize dictionaries to store favorability counts and percentages
char_fan_favorability = {}
char_fan_favorability_percentage = {}
char_no_fan_favorability = {}
char_no_fan_favorability_percentage = {}

#calculate fan ratings for each character and percentage of 'very favorably' ratings
for name in char_list:
    fan_chars = fan[name].value_counts()
    char_fan_favorability[name] = fan_chars['Very favorably']
total_fan_responses = len(fan.index)
for name in char_fan_favorability:
    char_fan_favorability_percentage[name] = round((char_fan_favorability[name]/total_fan_responses) * 100, 2)

#print favorability ratings
print('Percentage of Very Favorably ratings from fans:')    
for key, value in char_fan_favorability_percentage.items():
    print (key,value)

print('\n')
#calculate non-fan ratings for each character and percentage of 'very favorably' ratings
for name in char_list:
    no_fan_chars = no_fan[name].value_counts()
    char_no_fan_favorability[name] = no_fan_chars['Very favorably']
total_no_fan_responses = len(no_fan.index)
for name in char_no_fan_favorability:
    char_no_fan_favorability_percentage[name] = round((char_no_fan_favorability[name]/total_no_fan_responses) * 100, 2)

#print favorability ratings
print('Percentage of Very Favorably ratings from non-fans:')    
for key, value in char_no_fan_favorability_percentage.items():
    print (key,value)

#plot percentage of fan and non-fan favorability ratings
fig, ax = plt.subplots(figsize=(20,10))
bar_positions = np.arange(len(char_names))
ax.barh(bar_positions - width/2, char_fan_favorability_percentage.values(), width, label='Fan')
ax.barh(bar_positions + width/2, char_no_fan_favorability_percentage.values(), width, label='Not a Fan')
ax.set_yticks(bar_positions)
ax.set_yticklabels(char_fan_favorability_percentage)
ax.set_xlabel('Percentage of Respondents Who Rate the Character Very Favorably')
ax.legend()
ax.set_title('Very Favorably Ratings')
```

Perhaps unsurprisingly, non-fans had fewer 'Very favorably' ratings for every character. It is somewhat telling however, that the non-fan responses tended to be higher for the characters that are much more easily recognized such as Luke Skywalker, Han Solo, and Yoda. This is likely due to simple name-recognition, as these characters have become pop-culture icons independent of the Star Wars franchise. 

Fan favorites are predictable, with Han Solo taking the top spot, followed by Yoda and Obi-Wan Kenobi. More interesting is R2-D2, who captures the fourth ranking ahead of even Luke Skywalker and Princess Leia. 

It is also interesting that among fans and non-fans alike, villains Anakin and Darth Vader (yes, they're the same person) rank higher than film heroes Padme Amidala, Lando Calrissian, and Jar-Jar Binks. 

Almost no one likes Jar-Jar.
